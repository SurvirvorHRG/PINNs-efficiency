```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load Excel files
file1 = "SDGD_PINN_ref_Poisson_dim=10_batch=1_N_f=200_method=3_SEED=0_Num_params=34561.xlsx"  # Replace with actual file path
file2 = "SADGD_PINN_Poisson_dim=10_batch=1_N_f=200_method=3_SEED=0_Num_params=34561.xlsx"  # Replace with actual file path

df1 = pd.read_excel(file1)
df2 = pd.read_excel(file2)

# Extract relevant columns
epochs = range(1, len(df1) + 1)
metrics = ["loss", "L2", "L1"]

# Create subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle("Model Convergence Comparison")

for i, metric in enumerate(metrics):
    axes[i].semilogy(epochs, df1[metric], label="SDGD PINN")
    axes[i].semilogy(epochs, df2[metric], label="SADGD PINN")
    axes[i].set_xlabel("Epoch")
    axes[i].set_ylabel(metric)
    axes[i].set_title(f"{metric} over Epochs (Log Scale)")
    axes[i].legend()
    axes[i].grid(True, which="both", linestyle="--")

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
